```python
from transformers import pipeline
import pandas as pd
```

### Text Classification

```python
classifier = pipeline('text-classification')
text = "I love you !"
outputs = classifier(text)
pd.DataFrame(outputs)
```

## NER Tagger

```python
ner_tagger = pipeline("ner", aggregation_strategy="simple", model="dslim/bert-base-NER")
text = "This is Jonathan from Amazon Jungle"
outputs = ner_tagger(text)
pd.DataFrame(outputs)
```

## Question and Answering

```python
text = """
  Dear Amazon Support Team - Wrong product assigned to me,
  I order product A got product B instead :(. Please help !
"""
reader = pipeline("question-answering")

question = "Something wrong"

outputs = reader(question=question, context= text)

pd.DataFrame([outputs]).loc[0, 'answer']
```

# Summarization

```python
summarizer = pipeline('summarization')

outputs = summarizer(text)

outputs[0]["summary_text"]
```

# Text Generation

```python
from transformers import set_seed

set_seed(0)

generator = pipeline("text-generation")

response = "I am sorry to hear that your order was mixed up"

prompt = "user: " + text + " Customer Service Reponse " + response

outputs = generator(prompt, max_length=128)

outputs
```

# Translation

```python
translator = pipeline("translation_en_to_de", model="google-t5/t5-small")

outputs = translator(text)

outputs
```
